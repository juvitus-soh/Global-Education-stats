Visualizations

```python
import matplotlib.pyplot as plt
import seaborn as sns

#Distribution of Proficiency Levels at the End of Primary Education for english and math
plt.figure(figsize=(10, 6))
sns.histplot(data['Primary_End_Proficiency_Reading'], bins=20, kde=True, color='blue', label='Reading')
sns.histplot(data['Primary_End_Proficiency_Math'], bins=20, kde=True, color='orange', label='Math')
plt.title('Distribution of Proficiency Levels at the End of Primary Education. English and Math')
plt.xlabel('Proficiency Levels')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
#literacy rates per country and gender
plt.figure(figsize=(30, 18))
plt.bar(data['Countries and areas'], data['Youth_15_24_Literacy_Rate_Male'], label='Male')
plt.bar(data['Countries and areas'], data['Youth_15_24_Literacy_Rate_Female'], label='Female')
plt.xticks(rotation=90)
plt.xlabel('Countries and areas')
plt.ylabel('Literacy Rate')
plt.title('Literacy Rates by Country')
plt.legend()
plt.show()
```

```python
#comparing literacy rate between men and women in each country
literacy_rates = data[['Countries and areas', 'Youth_15_24_Literacy_Rate_Male', 'Youth_15_24_Literacy_Rate_Female']]
literacy_rates.set_index('Countries and areas', inplace=True)

literacy_rates
```

```python
#comparing literacy rate between men and women in each the world
average_male_literacy_rate = data['Youth_15_24_Literacy_Rate_Male'].mean()
average_female_literacy_rate = data['Youth_15_24_Literacy_Rate_Female'].mean()

print("Global Average Literacy Rates")
print("Male: {:.2f}%".format(average_male_literacy_rate))
print("Female: {:.2f}%".format(average_female_literacy_rate))
```

```python
#Exploit correlation
plt.figure(figsize=(24, 16))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#pairplot of some selected variables
sns.pairplot(data[['Gross_Primary_Education_Enrollment', 'Gross_Tertiary_Education_Enrollment', 'Birth_Rate']])
plt.suptitle('Pair Plot of Enrollment Rates and Birth Rate', y=1.02)
plt.show()
```

```python
#global youth literacy
data['Global_Youth_Literacy'] = data[['Youth_15_24_Literacy_Rate_Male', 'Youth_15_24_Literacy_Rate_Female']].mean(axis=1)
```

```python
#Visualizing the global literacy on the world map
import geopandas as gpd
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

#merge data 
merged_data = world.merge(data, how='left', left_on='name', right_on='Countries and areas')

#plot a choropleth map
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
merged_data.plot(column='Global_Youth_Literacy', cmap='Blues', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)
ax.set_title('Global_Youth_Literacy by Country')
plt.show()
```

Function to get some information about a specific country

```python
def get_country_details():
    country_name = input("Enter the name of a country: ")

    #filter the DataFrame to get details of specified country
    country_data = data[data['Countries and areas'] == country_name]

    # Check if the country exists in the dataset
    if country_data.empty:
        print("Country not found. Try making the initials in caps")
    else:
        print("\nCountry Details:")
        print("Country Name:", country_data['Countries and areas'].values[0])
        print("Male Literacy Rate:", country_data['Youth_15_24_Literacy_Rate_Male'].values[0])
        print("Female Literacy Rate:", country_data['Youth_15_24_Literacy_Rate_Female'].values[0])
        print("Primary Age Male Enrollment:", country_data['OOSR_Primary_Age_Male'].values[0])
        print("Primary Age Female Enrollment:", country_data['OOSR_Primary_Age_Female'].values[0])

#function Call
get_country_details()
```
